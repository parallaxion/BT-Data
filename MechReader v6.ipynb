```python
#load subset of data for more rapid testing

directory = 'mechs'
#directory = 'mechtest'
pd.set_option('display.max_colwidth', -1)
```

```python
weaponlist = {}
filecount = 0
mechs = {}
for item in os.listdir(directory):
    print(item)
    if os.path.isdir(os.path.join(directory,item)):
        #print("isdir")
    


        
        for filename in os.listdir(os.path.join(directory,item)):
            #print(filename)
            x = open(os.path.join(directory,item,filename), 'r')
            filename, file_extension = os.path.splitext(filename)
            mechs[filename] = {}
            #initialize data stuctures
            mechs[filename]["Speeds"] = {}
            #header=False 
            lc = 0
            readingweapons = False
            readingarmor = False
            for ln in x:
                lc = lc +1 #always count line number

                #reset certain values on iteration that is a blank line
                if ln in ['\n', '\r\n']:
                    readingweapons = False
                    if readingarmor == True:
                        #done summing armor, add total
                        #print(armorsum)
                        mechs[filename]["Armor Value"] = armorsum
                    readingarmor = False
                    continue
                    #move on to next line

                #remove line return character if continueing
                ln = ln.rstrip()        

                m =  mechs[filename]


                if ln.find("Version:") > -1:
                    #useless info, to me, skip it
                    continue
                    #mechs[filename]['Version'] = ln[8:]
                    #sub = True

                #line 2 is name
                if lc == 2:
                    mechs[filename]["Name"] = ln
                #line 3 is variant... etc
                if lc == 3: 
                    mechs[filename]["Variant"] = ln
                if lc == 5:
                    #syntax for string by index string[start:end], if end is empty [3:], its the last char
                    mechs[filename]["Config"] = ln[7:] 

                #parse by line content instead of line number necissary for most...
                #below: >-1 just means it found something, the return value is the strings index where the value starts!
                if ln.find("TechBase:") >-1:
                    mechs[filename]["TechBase"] = ln[9:]

               

                #armor parsing | complex
                if ln.find("Armor:",0,6) >-1:
                    mechs[filename]["Armor"] = ln[6:] #[ln[6:], []]
                    mechs[filename]["Armor Locations"]= []
                    readingarmor = True
                    armorsum = 0
                    continue
                    #example
                    #mechs[filename]["Weapons"][1].append("erppc")
                if readingarmor == True:
                    mechs[filename]["Armor Locations"].append(ln)
                    pos = ln.find(":") +1
                    armorsum = int(ln[pos:]) + armorsum


                #mass
                if ln.find("Mass:") > -1:
                     mechs[filename]["Mass"] = int(ln[5:])
                #year
                if ln.find("Era:") > -1: 
                    mechs[filename]["Era"] = int(ln[4:])
                #publication "source"
                if ln.find("Source:") > -1:
                    mechs[filename]["Source"] = ln[7:]
                #"rules level" aka complexity of play
                if ln.find("Rules Level:") > -1:
                    mechs[filename]["Rules Level"] = ln[12:]
                if ln.find("Engine:") > -1:
                    #in this case the engine size is up to a 3 digit number, min 2, its a q'n'd hack to capture the trailing space
                    # and convert the whole thing to int
                    mechs[filename]["Engine Size"] = int(ln[7:10]) 
                    mechs[filename]["Engine Type"] = ln[10:]
                if ln.find("Structure") > -1:
                    mechs[filename]["Structure"] = ln[10:]
                if ln.find("Myomer:") > -1:
                    mechs[filename]["Myomer"] = ln[7:]
                if ln.find("Heat Sinks:") > -1:
                    mechs[filename]["Heat Sinks"] = ln[11:]
                if ln.find("Walk MP:") > -1:
                    walk = int(ln[8:])
                    run = math.ceil(int(ln[8:]) * 1.5)
                    #mechs[filename]["Walk MP"] = walk
                    #print(math.ceil(int(ln[8:]) * 1.5))
                    #mechs[filename]["Run MP"] = run
                    mechs[filename]["Speeds"]["Walk"] = walk
                    mechs[filename]["Speeds"]["Run"] = run
                if ln.find("Jump MP:") > -1:
                    mechs[filename]["Speeds"]["Jump"] = ln[8:]
                filecount = filecount +1  
               
                #weapons parsing | complex
                if ln.find("Weapons") >-1:
                    mechs[filename]["Weapons"] = [ln[8:], []]
                    readingweapons = True
                    continue
                    #example
                    #mechs[filename]["Weapons"][1].append("erppc")
                if readingweapons == True:
                    mechs[filename]["Weapons"][1].append(ln)
                    try:
                        
                        wpncnt = int(ln[0:2])
                        weaponname = ln[2: int(ln.find(","))]
                        #print(wpncnt)
                        #print("more than one")
                    except:
                        #print("just one")
                        wpncnt = 1
                        weaponname = ln[0: int(ln.find(","))]
                    if weaponname in weaponlist:
                        weaponlist[weaponname] = weaponlist[weaponname] + wpncnt
                    else:
                        weaponlist[weaponname] = wpncnt
                #print(ass[8:]) #artifact from early testing, I have coding tourettes and all variable must be curse words in testing.
                #print(mechs[filename]['Version'])
    else:
        continue
#units parsed        

print(filecount)
print(weaponlist)
```
